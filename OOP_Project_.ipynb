# OOP with Banking System

```python
class Balance_Exception(Exception):
    pass

class BankAccount:
    def __init__(self,amt,ac_name):
        self.amt = amt
        self.ac_name = ac_name
        print(f"Hello {ac_name} your Account created \n Account Balance {amt:.2f}")

    def get_Balance(self):
        print(f"Account '{self.ac_name}' \nBalance = $ '{self.amt}'")

    def deposite(self, amount):
        self.amt += amount
        print("Deposite complete.")
        self.get_Balance()
        
    def validation_check(self,balance):
        if self.amt >= balance:
            return
        else:
            raise Balance_Exception(f"\nSorry, Account Don't have sufficient balance \nAvailable Amount $ '{self.amt}'")
    
    def withdraw_amt(self,withdraw):
        try:
            self.validation_check(withdraw)
            self.amt -= withdraw
            self.get_Balance()
        except Balance_Exception as error:
            print(error)

    def transfer(self,account,amount):
        try:
            print("==============\n Transfer Begien ...üöÄ")
            self.validation_check(amount)
            self.withdraw_amt(amount)
            account.deposite(amount)
            print("==============\n Transfer Completed ‚úÖ")

        except Balance_Exception as error:
            print("Transfer Interrupted ‚ùå ", error)

class InterestRewardsAcct(BankAccount):
    def deposite(self, amount):
        self.amt += (amount * 1.05)
        print("Deposite Compelete")
        self.get_Balance()


class Saving_Account(InterestRewardsAcct):
    def __init__(self, amt, ac_name):
        super().__init__(amt, ac_name)
        self.fees = 5

    def withdraw_amt(self, withdraw):
        try: 
            self.validation_check(withdraw + self.fees)
            self.amt -= (withdraw + self.fees)
            print("Withdraw Completed\n")
            self.get_Balance()
            print(f'\nProcessing Fee : $ {self.fees}')
        except Balance_Exception as error:
            print(f'withdraw Error: {error}')
```

```python
import tkinter as tk
from tkinter import messagebox, simpledialog

# Bank Account classes
class Balance_Exception(Exception):
    pass

class BankAccount:
    def __init__(self, amt, ac_name):
        self.amt = amt
        self.ac_name = ac_name

    def get_Balance(self):
        return self.amt

    def deposite(self, amount):
        self.amt += amount

    def validation_check(self, balance):
        if self.amt >= balance:
            return
        else:
            raise Balance_Exception(f"Sorry, Account doesn't have sufficient balance.\nAvailable Amount: ${self.amt}")

    def withdraw_amt(self, withdraw):
        try:
            self.validation_check(withdraw)
            self.amt -= withdraw
            return True
        except Balance_Exception as error:
            messagebox.showerror("Withdraw Error", str(error))
            return False

    def transfer(self, account, amount):
        try:
            self.validation_check(amount)
            self.withdraw_amt(amount)
            account.deposite(amount)
            messagebox.showinfo("Transfer Complete", f"Transferred ${amount} to account '{account.ac_name}'!")
            return True
        except Balance_Exception as error:
            messagebox.showerror("Transfer Error", str(error))
            return False

# App setup
class BankApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Bank Account System")
        self.root.geometry("400x500")
        self.accounts = {}

        # Create main frame
        main_frame = tk.Frame(root, padx=20, pady=20)
        main_frame.pack(fill=tk.BOTH, expand=True)

        # Title
        title_label = tk.Label(main_frame, text="üè¶ Bank Account System", font=("Arial", 16, "bold"))
        title_label.pack(pady=(0, 20))

        # Account creation section
        create_frame = tk.LabelFrame(main_frame, text="Create New Account", padx=10, pady=10)
        create_frame.pack(fill=tk.X, pady=(0, 20))

        tk.Label(create_frame, text="Account Name:").grid(row=0, column=0, sticky="w", pady=2)
        self.name_entry = tk.Entry(create_frame, width=20)
        self.name_entry.grid(row=0, column=1, padx=(10, 0), pady=2)

        tk.Label(create_frame, text="Initial Balance:").grid(row=1, column=0, sticky="w", pady=2)
        self.balance_entry = tk.Entry(create_frame, width=20)
        self.balance_entry.grid(row=1, column=1, padx=(10, 0), pady=2)

        create_btn = tk.Button(create_frame, text="Create Account", command=self.create_account, bg="#4CAF50", fg="white")
        create_btn.grid(row=2, column=0, columnspan=2, pady=(10, 0))

        # Account operations section
        operations_frame = tk.LabelFrame(main_frame, text="Account Operations", padx=10, pady=10)
        operations_frame.pack(fill=tk.X)

        # Account selection
        tk.Label(operations_frame, text="Select Account:").pack(anchor="w")
        self.selected = tk.StringVar(root)
        self.selected.set("No accounts created")
        
        # Initialize with placeholder - THIS FIXES THE ERROR
        self.account_menu = tk.OptionMenu(operations_frame, self.selected, "No accounts created")
        self.account_menu.pack(fill=tk.X, pady=(5, 10))

        # Operation buttons
        button_frame = tk.Frame(operations_frame)
        button_frame.pack()

        buttons = [
            ("üí∞ Check Balance", self.check_balance, "#2196F3"),
            ("üíµ Deposit", self.deposit, "#4CAF50"),
            ("üí∏ Withdraw", self.withdraw, "#FF9800"),
            ("üîÑ Transfer", self.transfer, "#9C27B0")
        ]

        for i, (text, command, color) in enumerate(buttons):
            btn = tk.Button(button_frame, text=text, command=command, 
                           width=15, bg=color, fg="white", pady=5)
            btn.grid(row=i//2, column=i%2, padx=5, pady=5)

        # Status display
        self.status_label = tk.Label(main_frame, text="Create an account to get started!", 
                                   fg="gray", font=("Arial", 10))
        self.status_label.pack(pady=(20, 0))

    def update_status(self, message):
        self.status_label.config(text=message)

    def refresh_menu(self):
        # Clear the menu
        self.account_menu['menu'].delete(0, 'end')
        
        # Add accounts to menu
        if self.accounts:
            for acc_name in self.accounts.keys():
                self.account_menu['menu'].add_command(
                    label=acc_name, 
                    command=tk._setit(self.selected, acc_name)
                )
            # Set first account as selected
            first_account = list(self.accounts.keys())[0]
            self.selected.set(first_account)
        else:
            self.selected.set("No accounts created")

    def create_account(self):
        name = self.name_entry.get().strip()
        balance_text = self.balance_entry.get().strip()
        
        if not name:
            messagebox.showerror("Input Error", "Please enter an account name!")
            return
            
        if not balance_text:
            messagebox.showerror("Input Error", "Please enter an initial balance!")
            return
            
        try:
            balance = float(balance_text)
            if balance < 0:
                messagebox.showerror("Input Error", "Initial balance cannot be negative!")
                return
        except ValueError:
            messagebox.showerror("Input Error", "Invalid balance amount! Please enter a number.")
            return
            
        if name in self.accounts:
            messagebox.showerror("Error", "Account already exists!")
            return
            
        # Create account
        acc = BankAccount(balance, name)
        self.accounts[name] = acc
        
        # Clear input fields
        self.name_entry.delete(0, tk.END)
        self.balance_entry.delete(0, tk.END)
        
        # Update GUI
        self.refresh_menu()
        self.update_status(f"Account '{name}' created successfully!")
        
        messagebox.showinfo("Success", f"Account '{name}' created with balance ${balance:.2f}")

    def get_selected_account(self):
        name = self.selected.get()
        if name == "No accounts created" or name not in self.accounts:
            messagebox.showerror("Error", "Please select a valid account!")
            return None
        return self.accounts[name]

    def check_balance(self):
        acc = self.get_selected_account()
        if acc:
            balance = acc.get_Balance()
            self.update_status(f"Balance for {acc.ac_name}: ${balance:.2f}")
            messagebox.showinfo("Balance", f"Account: {acc.ac_name}\nBalance: ${balance:.2f}")

    def deposit(self):
        acc = self.get_selected_account()
        if acc:
            amount = simpledialog.askfloat("Deposit", "Enter deposit amount:", minvalue=0.01)
            if amount is not None:
                acc.deposite(amount)
                new_balance = acc.get_Balance()
                self.update_status(f"Deposited ${amount:.2f} to {acc.ac_name}")
                messagebox.showinfo("Success", f"Deposited ${amount:.2f}\nNew Balance: ${new_balance:.2f}")

    def withdraw(self):
        acc = self.get_selected_account()
        if acc:
            amount = simpledialog.askfloat("Withdraw", "Enter withdrawal amount:", minvalue=0.01)
            if amount is not None:
                if acc.withdraw_amt(amount):
                    new_balance = acc.get_Balance()
                    self.update_status(f"Withdrew ${amount:.2f} from {acc.ac_name}")
                    messagebox.showinfo("Success", f"Withdrew ${amount:.2f}\nNew Balance: ${new_balance:.2f}")

    def transfer(self):
        acc = self.get_selected_account()
        if acc:
            if len(self.accounts) < 2:
                messagebox.showerror("Error", "Need at least 2 accounts for transfer!")
                return
                
            # Get target account
            target_options = [name for name in self.accounts.keys() if name != acc.ac_name]
            
            # Create selection dialog
            transfer_window = tk.Toplevel(self.root)
            transfer_window.title("Transfer Funds")
            transfer_window.geometry("300x200")
            transfer_window.transient(self.root)
            transfer_window.grab_set()
            
            tk.Label(transfer_window, text=f"Transfer from: {acc.ac_name}", font=("Arial", 12, "bold")).pack(pady=10)
            
            tk.Label(transfer_window, text="To Account:").pack()
            target_var = tk.StringVar(transfer_window)
            target_var.set(target_options[0])
            target_menu = tk.OptionMenu(transfer_window, target_var, *target_options)
            target_menu.pack(pady=5)
            
            tk.Label(transfer_window, text="Amount:").pack(pady=(10, 0))
            amount_entry = tk.Entry(transfer_window)
            amount_entry.pack(pady=5)
            amount_entry.focus()

            def do_transfer():
                try:
                    amount = float(amount_entry.get())
                    if amount <= 0:
                        raise ValueError("Amount must be positive")
                    target_acc = self.accounts[target_var.get()]
                    if acc.transfer(target_acc, amount):
                        self.update_status(f"Transferred ${amount:.2f} from {acc.ac_name} to {target_acc.ac_name}")
                    transfer_window.destroy()
                except ValueError:
                    messagebox.showerror("Error", "Please enter a valid amount!")
            
            # btn_frame = tk.Frame(transfer_window)
            # btn_frame.pack(pady=20)
            
            tk.Button(transfer_window, text="Transfer", command=do_transfer, bg="#4CAF50", fg="white").pack(side=tk.LEFT, padx=5)
            tk.Button(transfer_window, text="Cancel", command=transfer_window.destroy, bg="#f44336", fg="white").pack(side=tk.LEFT, padx=5)

# Run the application
if __name__ == "__main__":
    root = tk.Tk()
    app = BankApp(root)
    root.mainloop()
```
